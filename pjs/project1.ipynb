```python
# 1.평일과 주말간 대중교통 이용인원 비교 (ranksum test)
print(stats.ranksums(data_weekday['강남구전체이용인원'],data_weekend['강남구전체이용인원']))
```

```python
# 2.평일과 주말간 대중교통 이용률 비교 (normality test)
print(stats.shapiro(data_weekday['강남구버스이용률']))
print(stats.shapiro(data_weekend['강남구버스이용률']))
```

```python
# 2.평일과 주말간 대중교통 이용률 비교 (equal variance test)
print(stats.levene(data_weekday['강남구버스이용률'],data_weekend['강남구버스이용률']))
```

```python
# 2.평일과 주말간 대중교통 이용률 비교 (ranksum test)
print(stats.ranksums(data_weekday['강남구버스이용률'],data_weekend['강남구버스이용률']))
```

```python
# 3.평일 비가 올 때와 아닐 때의 대중교통 이용인원 비교 (normality test)
print(stats.shapiro(data_weekday_rain['강남구전체이용인원']))
print(stats.shapiro(data_weekday_norain['강남구전체이용인원']))
```

```python
# 3.평일 비가 올 때와 아닐 때의 대중교통 이용인원 비교 (equal variance test)
print(stats.levene(data_weekday_rain['강남구전체이용인원'],data_weekday_norain['강남구전체이용인원']))
```

```python
# 3.평일 비가 올 때와 아닐 때의 대중교통 이용인원 비교 (ranksum test)
print(stats.ranksums(data_weekday_rain['강남구전체이용인원'],data_weekday_norain['강남구전체이용인원']))
```

```python
# 4.평일 비가 올 때와 아닐 때의 대중교통 이용률 비교 (normality test)
print(stats.shapiro(data_weekday_rain['강남구버스이용률']))
print(stats.shapiro(data_weekday_norain['강남구버스이용률']))
```

```python
# 4.평일 비가 올 때와 아닐 때의 대중교통 이용률 비교 (equal variance test)
print(stats.levene(data_weekday_rain['강남구버스이용률'],data_weekday_norain['강남구버스이용률']))
```

```python
# 4.평일 비가 올 때와 아닐 때의 대중교통 이용률 비교 (ranksum test)
print(stats.ranksums(data_weekday_rain['강남구버스이용률'],data_weekday_norain['강남구버스이용률']))
```

```python
# 5.주말 비가 올 때와 아닐 때의 대중교통 이용인원 비교 (normality test)
print(stats.shapiro(data_weekend_rain['강남구전체이용인원']))
print(stats.shapiro(data_weekend_norain['강남구전체이용인원']))
```

```python
# 5.주말 비가 올 때와 아닐 때의 대중교통 이용인원 비교 (equal variance test)
print(stats.bartlett(data_weekend_rain['강남구전체이용인원'],data_weekend_norain['강남구전체이용인원']))
```

```python
# 5.주말 비가 올 때와 아닐 때의 대중교통 이용인원 비교 (ranksum test)
print(stats.ttest_ind(data_weekend_rain['강남구전체이용인원'],data_weekend_norain['강남구전체이용인원']))
```

```python
# 6.주말 비가 올 때와 아닐 때의 대중교통 이용률 비교 (normality test)
print(stats.shapiro(data_weekend_rain['강남구버스이용률']))
print(stats.shapiro(data_weekend_norain['강남구버스이용률']))
```

```python
# 6.주말 비가 올 때와 아닐 때의 대중교통 이용률 비교 (equal variance test)
print(stats.levene(data_weekend_rain['강남구버스이용률'],data_weekend_norain['강남구버스이용률']))
```
